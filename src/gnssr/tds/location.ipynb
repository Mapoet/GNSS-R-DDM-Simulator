```python
from sympy import *
init_printing()

x, y, h, tau, gamma, f, v_ty, v_tz, v_rx,v_ry, v_rz = symbols('x y h tau gamma f v_ty v_tz v_rx v_ry v_rz')

eq_delay = Eq(-tau + sqrt(x**2+(y+h/tan(gamma))**2+h**2)-h/sin(gamma)-y*cos(gamma))
eq_doppler = Eq(-f -v_ty*cos(gamma) - v_tz*sin(gamma) + (v_rx*x+v_ry*(y+h/tan(gamma))-v_rz*h)/sqrt(x**2+(y+h/tan(gamma))**2+h**2))
```

```python
eq_doppler
```

```python
solve([eq_delay,eq_doppler],[x,y])
```
